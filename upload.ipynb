```python
import boto3
import pandas as pd
import urllib.request
from bs4 import BeautifulSoup
import ssl
```

```python
url = 'https://sports-statistics.com/sports-data/college-football-stats-datasets-csv-database/'
```

```python
context = ssl.create_default_context()
context.check_hostname = False
context.verify_mode = ssl.CERT_NONE
```

```python
html = urllib.request.urlopen(url, context=context).read()
```

```python
# Use BeautifulSoup to parse the HTML
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Find all the anchor tags in the HTML
links = soup.find_all('a')
```

```python
s3 = boto3.client('s3')
bucket = s3.create_bucket(Bucket='bw-sports-analytics-data')
```

```python
for link in links:

    link_text = link.get('href')

    if link_text[-4:] == '.csv':
        lt_array = link_text.split('/')

        file_csv = lt_array[len(lt_array) - 1]
        conf = lt_array[len(lt_array)-2].replace('-', '')
        year = lt_array[len(lt_array)-3]

        file_name = file_csv[0:-4] + '-' + conf + '-' + year + ".parquet"
        print(file_name)
```
